```python
import networkx as nx
import pickle
import math
import matplotlib.pyplot as plt
```

```python
G = nx.read_gml("new_1000k_kzz=0.gml")
```

```python
iterations = 200; kfactor = 1.5; seed=5
pos = nx.fruchterman_reingold_layout(G, iterations=iterations, k=kfactor, threshold=1e-16, weight='weight', seed=seed)
```

```python
nx.draw_networkx_nodes(G, pos, alpha=1, node_color='#99bde0');
nx.draw_networkx_edges(G, pos, width=1, alpha=1, edge_color='#5b5f6d');
plt.axis('off');
```
