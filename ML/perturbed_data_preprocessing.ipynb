```python
for spread in ["50"]: # There is no data for other uncertainties
    for removed_species in ["CH4", "CO", "H2O", "NH3"]:
        for file_name in os.listdir(original_data_dir + "%s_removed/"%removed_species):
            print(file_name)        
            kzz_string = file_name.split(" kzz")[0]
        
            ave_temp = file_name.split("observables ")[1].split("k.csv")[0]
            # print(kzz_label[kzz_string], type(kzz_label[kzz_string]), kzz_legend[kzz_string])
        
            df = pd.read_csv(original_data_dir + "%s_removed/"%removed_species + file_name)
        
            df_new = df.drop("Unnamed: 0", axis=1)
        
            df_new["Kzz"] = kzz_label[kzz_string]
            df_new["%s abundance"%removed_species] = 0
            
            df_new.to_csv(result_dir + "%s_removed/"%removed_species + 
                "%s_removed_%sk_spread_kzz%s-aveT%s"%(removed_species, spread, kzz_label[kzz_string], ave_temp), index=False)

```
